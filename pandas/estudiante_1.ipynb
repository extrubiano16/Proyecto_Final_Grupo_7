```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
def suma_digitos(valor):
    return sum(int(digito) for digito in str(valor) if digito.isdigit())
```

```python
df['n'] = df['Número de documento'].apply(suma_digitos)
```

```python
def totall(n):
    x = 105
    tot = n * x**2 + 2 * x + 23
    
    x = 100
    tota = n * x**2 + 2 * x + 23
    
    total = (tot - tota) / 5
    
    return round(total, 2)
```

```python
df["P1"] = df["n"].apply(totall)
```

```python
df["Resultado-1"] = (df[df.columns[1]] == df["P1"]) * 5
```

```python
def C(x, n):
    return n * x**2 + 2*x + 23

def cp2(n):
    return round(C(101, n) - C(100, n), 2)

def cp3(n):   
    return round(200*n + 2, 2)
```

```python


df["P2"] = df["n"].apply(cp2)
```

```python
df = df.drop(columns=["CP2", "Calificacion2"], errors="ignore")
```

```python
df["Resultado-2"] = (df[df.columns[2]] == df["P2"]) * 5
```

```python
df["P3"] = df["n"].apply(cp3)
```

```python
df["Resultado-3"] = (df[df.columns[3]] == df["P3"]) * 5
```
