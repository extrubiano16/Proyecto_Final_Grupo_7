```python
import pandas as pd
```

```python
df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")
```

```python
def suma_digitos(valor):
    return sum(int(digito) for digito in str(valor) if digito.isdigit())
```

```python
df['n'] = df['Número de documento'].apply(suma_digitos)
```

```python
def totall(n):
    x = 105
    tot = n*x**2 + 2*x +23
    x = 100
    tota = n *x**2 + 2*x + 23
    total = (tot-tota)/5
    return round(total,2)
```

```python
df["P1"] = df["n"].apply(totall)
```

```python
df["Resultado-1"] = (df[df.columns[1]] == df["P1"]) * 5
```
